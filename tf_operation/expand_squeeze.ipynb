In tensorflow we can expland dimension :
1. tf.expand_dims(tf_tensor,axis=)
2. tf_tensor[tf.newaxis] 

In this notebook we will learn both of these methods

```python
import tensorflow as tf 
```

Create sample variables

```python
var1=tf.random.normal(shape=(5,)) ## 1 dimensional tensor
var2=tf.random.normal(shape=(2,5)) ## 2 dimensional tensor
var3= tf.random.normal(shape=(100,32,35))
```

#### expand_dims

```python
tf.expand_dims(var1,axis=0) ## Add a dimension in axis 0 
```

```python
tf.expand_dims(var1,axis=1) ## add a dimension in axis 1 (Guess the shape before running this cell)
```

```python
## See the change in 2 dimensional : var2.shape ==> 2,5
tf.expand_dims(var2,axis=0)
```

```python
tf.expand_dims(var2,axis=1)
```

```python
tf.expand_dims(var2,axis=2)
```

```python
## varible 3 ; orginal shape ==> 100,32,35
tf.expand_dims(var3,axis=0)
```

```python
tf.expand_dims(var3,axis=3) ## Add dimension in last axis
```

 Explanding with tf.newaxis

```python
# var1 shape=> 5
var1[tf.newaxis] ## if no slicing is given add a dims at axis =1
```

```python
var1[tf.newaxis,:] ## same as above(add dimension at axis =0, and leave other same)
```

```python
var1[:,tf.newaxis] ## Add dimension at the end 
```

```python
##with var2; original shape ==> 2,5
var2[tf.newaxis]
```

```python
var2[tf.newaxis,:,:]
```

```python
var2[:,tf.newaxis,:] ## So similar to tf.expand_dims. use slicing, add new tf.newaxis in that axis where you want to expand axs
```

```python
# I can also tf.newxis multiple times; this makes it simple to use
# for example you have a image with 2 dimension; h,w; since you want to train it you have to add batch size as well as channel=1
h,w=480,640
img_var=tf.random.normal(shape=(h,w))
## In order to predict the result from cnn model it should have batch size(new axis at 0) as well as channel(new axis at end)

img_var[tf.newaxis,:,:,tf.newaxis] ##  batchs_size,h,w,c; One liner solutions
```

## squeeze
This operation is useful to reduce dimension with value 1. It can be useful to remove the batch axis . Two ways to do it:
- tf.squeeze
- tensorflow slicing

```python
expanded_var=img_var[tf.newaxis,:,:,tf.newaxis]
```

```python
## squeenze or unexpand

reduced_batch=tf.squeeze(expanded_var,axis=0)
red_to_img=tf.squeeze(reduced_batch,axis=2)
```

```python
red_to_img
```

```python
## using slicing (one liner)

expanded_var[0,:,:,0] ## Put 0 to those where you want to reduce dimension 
```
